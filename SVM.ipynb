```python
from sklearn.datasets import make_blobs, make_circles
import matplotlib.pyplot as plt
from sklearn import svm
import numpy as np
```

```python
print('Сгенерируйте 2 линейно разделимых класса.')
X, y = make_blobs(centers=2)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
print('Примените SVM, визуализируйте разделяющую гиперплоскость.')

X, y = make_blobs(centers=2)
plt.scatter(X[:, 0], X[:, 1], c=y)

clf = svm.SVC(kernel='linear', C=1)
clf.fit(X, y)

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1])
yy = np.linspace(ylim[0], ylim[1])
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.show()
```

```python
print('Сгенерируйте не разделимую линейно выборку (но разделимую кривой второго порядка).')
X, y = make_circles(n_samples=100, noise=0.05, factor=.5)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
print('Попробуйте применить SVM к этой выборке.')
X, y = make_circles(n_samples=100, noise=0.05, factor=.5)
plt.scatter(X[:, 0], X[:, 1], c=y)

clf = svm.SVC(kernel='linear', C=1)
clf.fit(X, y)

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1])
yy = np.linspace(ylim[0], ylim[1])
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.show()
```

```python
print('Как можно построить разделяющую поверхность при помощи SVM? (придумайте, 2 способа, см. ядра (kernel))')
print('Визуализируйте результат.')

X, y = make_circles(n_samples=100, noise=0.05, factor=.5)
plt.scatter(X[:, 0], X[:, 1], c=y)

clf = svm.SVC(kernel='rbf', C=1, gamma=1)
clf.fit(X, y)

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1])
yy = np.linspace(ylim[0], ylim[1])
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.show()
```

```python
X, y = make_circles(n_samples=100, noise=0.05, factor=.5)
plt.scatter(X[:, 0], X[:, 1], c=y)

clf = svm.SVC(kernel='poly', gamma=1, degree=2)  # даст результат при четной степени
clf.fit(X, y)

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1])
yy = np.linspace(ylim[0], ylim[1])
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

plt.scatter(X[:, 0], X[:, 1], c=y)
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.show()
```
